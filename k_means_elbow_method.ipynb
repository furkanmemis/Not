```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Veri setini oluşturma
X = np.array([[1, 1], [2, 1], [4, 3], [5, 4], [6, 5]])

# Küme sayılarını ve SSE değerlerini saklamak için boş listeler
k_values = range(1, 5)
sse_values = []

# Her bir küme sayısı için k-means uygula ve SSE hesapla
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sse_values.append(kmeans.inertia_)
```

```python
# SSE değerlerini grafiğe çizdirme
plt.plot(k_values, sse_values, 'bx-')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```
